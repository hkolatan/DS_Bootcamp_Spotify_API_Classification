```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
sns.set_style("whitegrid")
sns.set_context("paper")

train_sizes, train_scores, test_scores = learning_curve(grid_search.best_estimator_, X_train, Y_train, cv=5, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10))

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.figure(figsize=(12, 7))
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training Accuracy")
plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Validation Accuracy")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve for KNN")
plt.legend(loc="best")
plt.savefig("Learning_Curve_for_KNN.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split

# Veri kümesini yükleyin veya oluşturun
# X ve y verilerini tanımlayın

# Train ve test veri setlerini oluşturun
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# KNN modelini tanımlayın
knn = KNeighborsClassifier()

# Komşu sayılarını belirleyin
neighbors = range(1, 50)

# Boş listeler oluşturun
precisions = []
recalls = []

# Her komşu sayısı için modeli eğitin ve precision, recall değerlerini kaydedin
for k in neighbors:
    knn.set_params(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_prob = knn.predict_proba(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
    precisions.append(precision)
    recalls.append(recall)

plt.figure(figsize=(12, 7))
for k in range(len(neighbors)):
    plt.plot(recalls[k], precisions[k], label='k={}'.format(neighbors[k]))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for Different k Values in KNN')
plt.legend(loc='lower center', ncol=5)
plt.tight_layout()
plt.savefig("DET_Curve.png", dpi = 300)
plt.show()
```

```python
def evaluate_model(model, X, Y, pred_label=1):

   y_pred = model.predict(X)

   accuracy = round(accuracy_score(Y, y_pred), 3)
   recall = round(recall_score(Y, y_pred, pos_label = pred_label), 3)
   precision = round(precision_score(Y, y_pred, pos_label = pred_label), 3)
   f1 = round(f1_score(Y, y_pred, pos_label = pred_label), 3)
   auc_score = round(roc_auc_score(Y, y_pred), 3)

   metric_df = pd.DataFrame([accuracy, recall, precision, f1, auc_score]).T
   metric_df.index = ["Class " + str(pred_label)]
   metric_df.columns = ["Accuracy", "Recall", "Precision", "F1 Score", "AUC Score"]

   return metric_df
```

```python
def make_confusion_matrix(model, threshold=0.5):
    
    y_predict = (model.predict_proba(X_test)[:, 1] >= threshold)
    fraud_confusion = confusion_matrix(Y_test, y_predict)
    
    plt.figure(dpi=80)
    sns.heatmap(fraud_confusion, cmap=plt.cm.Blues, annot=True, square=True, fmt='d',
           xticklabels=['Active', 'Relax'],
           yticklabels=['Active', 'Relax'])
    plt.xlabel('Prediction')
    plt.ylabel('Actual')
```
